 - Per√≠odo

```python
T = []
for i in range(1, len(robo['t'])):
    T.append(robo['t'][i] - robo['t'][i-1])

len(T)
```

```python
T = pd.DataFrame(T)
```

```python
# media 
Tmedia = np.mean(T)
Tmedia
```

```python
# variancia
np.var(T)
```

```python
# desvio padr√£o
np.std(T)
```

```python
# maximo
np.max(T)
```

```python
# minimo
np.min(T)
```

- Jitter

```python
Delta_T = T - Tmedia
Delta_T
```

```python
Jitter = np.mean(np.abs(Delta_T))
Jitter
```
